```python
# Dans 'sous_nutrition', remplacer les laps de temps [2012-2014], [2013-2015]... par '2013', '2014'...
sous_nutrition['Année'].replace(['2012-2014', '2013-2015', '2014-2016', '2015-2017', '2016-2018', '2017-2019'], [2013, 2014, 2015, 2016, 2017, 2018], inplace=True)

# Jointure entre la table 'population' et la table 'sous nutrition', en ciblant l'année 2017
personne_sn = pd.merge(population, sous_nutrition, on=['Année', 'Zone'], how ='left')
personne_sn = personne_sn.loc[personne_sn['Année'] == 2017, :]
personne_sn.head()
```

```python
#Affichage du dataset
print("Le tableau comporte {} observation(s) ou article(s)".format(personne_sn.shape[0]))
print("Le tableau comporte {} colonne(s)".format(personne_sn.shape[1]))
personne_sn.info()
```

```python
# Calculer la somme des personnes en sous nutrition et la population totale
Total_personnes_sn=personne_sn['sous_nutrition'].sum()
Total_population=personne_sn['Population'].sum()
# Calculer le proportion des personnes en état de sous nutrition
Ration_personnes_sn = (Total_personnes_sn * 100) / Total_population
# Affichage du nombre de personnes en état de sous nutrition
print('Personnes en état de sous nutrition :   ',f"{Total_personnes_sn:,}")
print('Population totale :                   ', f"{Total_population:,}")
print('Pourcentage des personnes en état de sous nutrition: ',round(Ration_personnes_sn,2),"%")

# Préparation graphique
Etatp = ["Population en sous-nutrition", "Population mondiale"]
PopuOk= round((Total_population-Total_personnes_sn),0)
Pop = [Total_personnes_sn, PopuOk]
df = pd.DataFrame(list(zip(Etatp,Pop)), columns = ['Gatégorie','Valeur'])
df['Valeur'] = df['Valeur'].astype("float")
# Conception graphique
fig = plt.figure(figsize =(4, 4))  
fig.set_facecolor('none')
plt.title("Proportion des personnes en sous nutrition - 2017", y=1.05, 
          fontdict={'size': 20, 'weight': 'bold', 'style':'italic', 'color': 'teal'})
plt.pie(df['Valeur'], explode=(0.3, 0), labels=df['Gatégorie'], colors=['firebrick', 'lightseagreen'], 
        autopct='%1.2f%%', textprops=dict(color="black", weight= 'bold'), shadow=True, startangle=0 )
# display graphique
plt.show()
```

### Suivant les informations recueillies sur le site [Vidal](https://www.vidal.fr/sante/nutrition/equilibre-alimentaire-adulte/recommandations-nutritionnelles-adulte.html),  je me suis basé sur un apport moyen                  par personne de 2,250 Kcal/jour

```python
#On commence par faire une jointure entre le data frame population et Dispo_alimentaire afin d'ajouter dans ce dernier la population
population2017 = population.loc[population['Année'] == 2017, :]
Dispo_alim = pd.merge(dispo_alimentaire, population2017, on=['Zone'], how ='left')
```

```python
Dispo_alim.head() #Affichage du nouveau dataframe
```

```python
#Création de la colonne dispo_kcal avec calcul des kcal disponibles mondialement
Dispo_alim['dispo kcal/jour'] = Dispo_alim["Disponibilité alimentaire (Kcal/personne/jour)"] * Dispo_alim['Population']
```

```python
# Calcul du nombre d'humains pouvant être nourris # dispo alimentaire totale en kcal 

# Calcul de la somme des dispositions alimentaires
Total_dispo_alim = Dispo_alim["dispo kcal/jour"].sum()
Total_dispo_alim_an = Dispo_alim["dispo kcal/jour"].sum() * 365
print('Les dispositions alimentaires sont de :      ', f"{Total_dispo_alim:,}", 'Kcal/jour')
print('Les dispositions alimentaires sont de :   ', f"{Total_dispo_alim_an:,}", 'Kcal/an')
print('-------------')
# Calcul du nombre de personnes pouvant être nourris et de sa proportion
Total_dispo_par_personne = round((Total_dispo_alim / 2250),0)
print('Population totale :                                     ', f"{Total_population:,}", 'personnes')
print('Les dispositions alimentaires permettent de nourrir:    ', f"{Total_dispo_par_personne:,}", 'personnes')
Ratio = round(((Total_dispo_par_personne * 100) / Total_population),2)
print('la proportion de personnes pouvant être nourris est de :', Ratio, "%")
```

```python
#Calcul de la disponibilité totale
som = (Dispo_alim['Disponibilité intérieure'].sum())
print('La disponibilité intérieure totale est de :  ', f"{som:,}", 'Kg')
```

```python
#création d'une boucle for pour afficher les différentes valeurs en fonction des colonnes aliments pour animaux, pertes, nourritures, 
Uti_Acq_dispo_alim = ['Aliments pour animaux', 'Pertes', 'Nourriture', 'Semences', 'Traitement', 'Autres Utilisations']

DisposInt = Dispo_alim['Disponibilité intérieure'].sum()
for i in Uti_Acq_dispo_alim:
    Dispo = (Dispo_alim[i].sum())
    print("Proportion de", i, ":", "{:.2f}".format(Dispo_alim[i].sum()*100/DisposInt), "% - Dispo : ",f"{Dispo:,}")

# Préparation des données du graphique   
Pourcentage = []
Gatégorie = []
DisposInt1 = Dispo_alim['Disponibilité intérieure'].sum()
plt.figure(figsize=(10, 5))
for i in Uti_Acq_dispo_alim:
    Pourcentage.append((round(Dispo_alim[i].sum()*100/DisposInt1,2)))
    Gatégorie.append(i)
df = pd.DataFrame(list(zip(Gatégorie,Pourcentage)), columns = ['Gatégorie','Pourcentage'])
df = df.sort_values(by=['Pourcentage'], ascending=False)
df.reset_index()
# Mise en page du Graphique
sns.set(rc={'axes.facecolor':'black', 'figure.facecolor':'white'})
fig = plt.figure(figsize=(12, 4))
plt.title("Utilisation de la disponibilité intérieure en 2017", y=1.05, 
          fontdict={'size': 20, 'weight': 'bold', 'style':'italic', 'color': 'teal'})
Pro = df['Pourcentage'].tolist()
for i in range(6):
    plt.text(Pro[i]+1, i+0.1, Pro[i], weight = 'bold', color = 'white')
# Génération du Graphique    
gfg = sns.barplot(x='Pourcentage',y='Gatégorie', data=df, palette="Greens");
gfg.set_xlim(0, 55)
plt.show()    
```

```python
# On calcule la proportion exportée en fonction de la production
Manioc_Thai = Dispo_Thai.loc[Dispo_Thai['Produit'] == 'Manioc', :]
Manioc_Thaip = round((( Manioc_Thai['Exportations - Quantité'] * 100 ) / Manioc_Thai['Production']),2)

Di = Manioc_Thai['Disponibilité intérieure'].sum()
Pq = Manioc_Thai['Production'].sum()
Iq = Manioc_Thai['Importations - Quantité'].sum()
Eq = Manioc_Thai['Exportations - Quantité'].sum()
Vs = Manioc_Thai['Variation de stock'].sum()

print('Proportion de Manioc exportée:', "{:.2f}".
      format(Manioc_Thai['Exportations - Quantité'].iloc[0]*100 /  Manioc_Thai['Production'].iloc[0]),
      "%")
print("Vérification de l'équation de Julien :  ", f"{Di:,}", "=", f"{Pq + Iq - Eq + Vs:,}")
print("Quantité exporté :  ", f"{Manioc_Thai['Exportations - Quantité'].sum():,}")
print("Quantité produite : ", f"{Manioc_Thai['Production'].sum():,}")

EQ_Thai = Manioc_Thai['Exportations - Quantité'].name, Manioc_Thai['Production'].name
P_Thai = Manioc_Thai['Exportations - Quantité'].values, (Manioc_Thai['Production']-
                    Manioc_Thai['Exportations - Quantité']).values
df = pd.DataFrame(list(zip(EQ_Thai,P_Thai)), columns = ['Gatégorie','Valeur'])
df['Valeur'] = df['Valeur'].astype("float")
# define Seaborn color palette to use
palette_color = sns.color_palette('Reds')
fig = plt.figure(figsize =(4, 4))  
fig.set_facecolor('none')
# plotting data on chart
plt.title("Utilisation de la production intérieure de Manioc", y=1.05, 
          fontdict={'size': 20, 'weight': 'bold', 'style':'italic', 'color': 'teal'})
plt.pie(df['Valeur'], explode=(0.3, 0), labels= ["Destiné à l'Export", 'Utilisation nationale'], colors=['firebrick', 'lightseagreen'], 
        autopct='%1.2f%%', textprops=dict(color="black", weight= 'bold'), shadow=True, startangle=0 )
# displaying chart
plt.show()
```
